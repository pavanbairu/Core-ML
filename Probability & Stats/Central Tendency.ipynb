```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Student_Name=  ["Adam", "Bella", "Charlie", "David", "Emily", "Frank", "Grace", "Harry", "Isabella", "Jack"]
Score = [56, 78, 90, 82, 60, 95, 72, 88, 80, 76]
```

```python
data = pd.DataFrame(
    {"Student_Names" : Student_Name,
     "scores" : Score
    })
data
```

```python
np.mean(data['scores']), np.median(data['scores']), data['scores'].mode()[0]
```

```python
import seaborn as sns
sns.histplot(data["scores"], kde=True, color="skyblue", bins=10)
```

```python
a = pd.DataFrame([10,20,20,30,40,50,50,50,60], columns=['values'])
a['values'].mode()
```

```python
data['scores'].mode()
```

```python
# Step 1: Calculate the Mean Score
sns.kdeplot(data["scores"], color="blue", fill=True, alpha=0.4, label="Score Density")
# Add vertical lines for mean
plt.axvline(np.mean(data['scores']), color='green', linestyle='--', linewidth=2, label=f'Mean: {np.mean(data['scores']):.2f}')

# Add legend and labels
plt.legend()
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.title("Distribution of Scores with Mean")

# Show plot
plt.show()
```

```python
sns.kdeplot(data["scores"], color="blue", fill=True, alpha=0.4, label="Score Density")
# Add vertical lines for median
plt.axvline(np.median(data['scores']), color='blue', linestyle='-', linewidth=2, label=f'Median: {np.median(data['scores']):.2f}')

# Add legend and labels
plt.legend()
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.title("Distribution of Scores with Median")

# Show plot
plt.show()
```

```python
sns.kdeplot(data["scores"], color="blue", fill=True, alpha=0.4, label="Score Density")
# Add vertical lines for mean and median
plt.axvline(np.mean(data['scores']), color='green', linestyle='--', linewidth=2, label=f'Mean: {np.mean(data['scores']):.2f}')
plt.axvline(np.median(data['scores']), color='blue', linestyle='-', linewidth=2, label=f'Median: {np.median(data['scores']):.2f}')

# Add legend and labels
plt.legend()
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.title("Distribution of Scores with Mean and Median")

# Show plot
plt.show()
```

```python
sns.kdeplot(data["scores"], color="blue", fill=True, alpha=0.4, label="Score Density")
# Add vertical lines for mean, median, and mode
plt.axvline(np.mean(data['scores']), color='green', linestyle='--', linewidth=2, label=f'Mean: {np.mean(data['scores']):.2f}')
plt.axvline(np.median(data['scores']), color='blue', linestyle='-', linewidth=2, label=f'Median: {np.median(data['scores']):.2f}')
plt.axvline((data['scores'].mode()[0]), color='red', linestyle='-.', linewidth=2, label=f'Lower Mode: {(data['scores'].mode()[0]):.2f}')

# Add legend and labels
plt.legend()
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.title("Distribution of Scores with Mean, Median, and Mode")

# Show plot
plt.show()
```

```python
sns.kdeplot(data["scores"], color="blue", fill=True, alpha=0.4, label="Score Density")
# Add vertical lines for mean, median, and mode
plt.axvline(np.mean(data['scores']), color='green', linestyle='--', linewidth=2, label=f'Mean: {np.mean(data['scores']):.2f}')
plt.axvline(np.median(data['scores']), color='blue', linestyle='-', linewidth=2, label=f'Median: {np.median(data['scores']):.2f}')
plt.axvline((data['scores'].mode()[9]), color='red', linestyle='-.', linewidth=2, label=f'Higher Mode: {(data['scores'].mode()[9]):.2f}')

# Add legend and labels
plt.legend()
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.title("Distribution of Scores with Mean, Median, and Mode")

# Show plot
plt.show()
```
